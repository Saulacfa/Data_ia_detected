```python
# Crear DataFrame con textos humanos (muestra)
data_list = [data1, data2, data4, data6, data8, data10, data12, data14, data16, data18, data20, data40, data80, data160]

muestras_procesadas = {}  # To store the processed samples

for i, df in enumerate(data_list):
    # Get the sample size from the original 'muestras' dictionary
    tamano = tamanos[i]

    # Create DataFrames for human and AI texts within this sample
    human_texts_df = pd.DataFrame({
        'text': df['human_text'],
        'generated': 0
    })
    ai_texts_df = pd.DataFrame({
        'text': df['ai_text'],
        'generated': 1
    })
    data_list[i] = pd.concat([human_texts_df, ai_texts_df])
    print(data_list[i])

```

```python
import os

# ... (your previous code to create data_list and tamanos) ...

# Create a directory to store the CSV files (if it doesn't exist)
output_dir = 'muestras_csv'
os.makedirs(output_dir, exist_ok=True)

# Loop to save each DataFrame in data_list to a separate CSV file
for i, df in enumerate(data_list):
    # Get the original sample size from the tamanos list
    tamano_original = tamanos[i]

    # Calculate the double sample size
    tamano_doble = tamano_original * 2

    # Construct the filename with the double sample size
    nombre_archivo = os.path.join(output_dir, f"data{tamano_doble}.csv")

    df.to_csv(nombre_archivo, index=False)
    print(f"Muestra guardada en: {nombre_archivo}")
```
